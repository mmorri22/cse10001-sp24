### C Code

    #include <stdio.h>
    #include <stdlib.h>

    void alloc_array( int** array, long unsigned int len ){

        fprintf( stdout, "%p %lu\n", &len, len);

        *array = (int *)malloc( len * sizeof(int) );

        fprintf( stdout, "%p %p\n", &array, array);
    }

    void print_stat( int array[], long unsigned int len ){

        fprintf( stdout, "%p %lu\n", &len, len);

        long unsigned int iter;
        for(iter = 0; iter < len; ++iter){
            fprintf( stdout, "%p %d\n", &array[iter], array[iter]);
        }

    }

    void update_dyn( int* array, long unsigned int len ){

        fprintf( stdout, "%p %lu\n", &len, len);
        fprintf( stdout, "%p %p\n", &array, array);

        long unsigned int iter;
        for(iter = 0; iter < len; ++iter){

            fprintf( stdout, "Enter an integer: ");
            if( fscanf( stdin, "%d", &array[iter]) ) {
                fprintf( stdout, "%p %d\n", &array[iter], array[iter]);
            }
        }

    }

    int main()
    {

        int static_array[5] = {18, 2, 10, 15, -25};

        print_stat( static_array, 5 );


        long unsigned int dyn_array_len = 6;
        fprintf( stdout, "%p %lu\n", &dyn_array_len, dyn_array_len);

        int* dynamic_array;
        alloc_array( &dynamic_array, dyn_array_len );
        
        fprintf( stdout, "%p %p\n", &dynamic_array, dynamic_array);


        update_dyn( dynamic_array, dyn_array_len);


        free( dynamic_array );

        return 0;
    }
